```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('data/baseResultsOverAlphaBeta.csv')
base['Dataset'] = 'SG-BASE'
base['Dataset Num'] = 0
hetero = pd.read_csv('data/heteroResultsOverAlphaBeta.csv')
hetero['Dataset'] = 'SG-HETEROPHILIC'
hetero['Dataset Num'] = 1
unfair = pd.read_csv('data/unfairResultsOverAlphaBeta.csv')
unfair['Dataset'] = 'SG-UNFAIR'
unfair['Dataset Num'] = 2
less = pd.read_csv('data/lessinformResultsOverAlphaBeta.csv')
less['Dataset'] = 'SG-LESSINFORM'
less['Dataset Num'] = 3
more = pd.read_csv('data/moreinformResultsOverAlphaBeta.csv')
more['Dataset'] = 'SG-MOREINFORM'
more['Dataset Num'] = 4
lst = [base, hetero, unfair, less, more]
df = pd.concat(lst)
df.head()
```

```python
alpha_consist = df[df['what is og'] == 'og alpha']
beta_consist = df[df['what is og'] == 'og beta']
```

```python
curr = alpha_consist[alpha_consist['Dataset'] == 'SG-BASE']
plt.scatter(list(curr['beta']), list(curr['best acc']), color='#5d0379', marker='o', label='Best Accuracy')
plt.scatter(list(curr['beta']), list(curr['best f1']), color='#bb87cc', marker='o', label='Best F1 Score')
plt.scatter(list(curr['beta']), list(curr['best faith']), color='#046d1d', marker='v', label='Best Unfaithfulness')
plt.title('Beta Parameter Sweep Over Consistant Alpha Parameter for SG-BASE')
plt.xlabel('Beta')
plt.ylabel('Results')
plt.legend()
```
