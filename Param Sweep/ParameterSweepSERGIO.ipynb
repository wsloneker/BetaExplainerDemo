```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df25 = pd.read_csv('data/25_SERGIOResultsOverAlphaBeta.csv')
df25['Dataset'] = 'SERGIO 25% Sparsity'
df50 = pd.read_csv('data/50_SERGIOResultsOverAlphaBeta.csv')
df50['Dataset'] = 'SERGIO 50% Sparsity'
lst = [df25, df50]
df = pd.concat(lst)
df.head()
```

```python
alpha_consist = df[df['what is og'] == 'og alpha']
beta_consist = df[df['what is og'] == 'og beta']
```

```python
curr = alpha_consist[alpha_consist['Dataset'] == 'SERGIO 25% Sparsity']
plt.scatter(list(curr['beta']), list(curr['full acc']), color='#5d0379', marker='o', label='Accuracy Including False Negatives')
plt.scatter(list(curr['beta']), list(curr['full f1']), color='#bb87cc', marker='o', label='F1 Score Including False Negatives')
plt.scatter(list(curr['beta']), list(curr['best faith']), color='#046d1d', marker='v', label='Best Unfaithfulness')
plt.title('Beta Parameter Sweep Over Consistant Alpha Parameter for SERGIO 25% Sparsity')
plt.xlabel('Beta')
plt.ylabel('Results')
plt.legend()
```

```python
curr = alpha_consist[alpha_consist['Dataset'] == 'SERGIO 25% Sparsity']
plt.scatter(list(curr['beta']), list(curr['partial acc']), color='#5d0379', marker='o', label='Accuracy Dis-cluding False Negatives')
plt.scatter(list(curr['beta']), list(curr['partial f1']), color='#bb87cc', marker='o', label='F1 Score Dis-cluding False Negatives')
plt.scatter(list(curr['beta']), list(curr['best faith']), color='#046d1d', marker='v', label='Best Unfaithfulness')
plt.title('Beta Parameter Sweep Over Consistant Alpha Parameter for SERGIO 25% Sparsity')
plt.xlabel('Beta')
plt.ylabel('Results')
plt.legend()
```
